（1）画出$y=x^{2}+2x+1$在区间[-5,3]的函数图像。

```python
t = np.linspace(-5.,3.,200)
#plt.axis([-5,5,-5,20])
plt.figure(figsize=(9,6))
plt.plot(t,t**2+t*2+1,'--')
plt.xlabel('x')
plt.ylabel('y')
plt.plot([-1,-1],[0,16],'r--')
plt.show()
```

（2）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.

```python
plt.figure(figsize = (10,6))
a = np.arange(-np.pi,np.pi,0.1)
plt.subplot(211)
plt.plot(a,np.sin(a),'b',linewidth = 2.5)

plt.subplot(212)
plt.plot(a,np.cos(a),"r--",linewidth = 2.5)
plt.show()
```

（3）读取上次作业保存的酒店数据，画出每个地区酒店数量的柱状图，柱状颜色为红色

```python
import pandas as pd
```

```python
df = pd.read_excel('酒店数据2.xlsx',index_col = 0)
df[:10]
```

```python
data = df["地区"].value_counts()
data
```

```python
x = data.index
y = data.values

plt.figure(figsize = (10,6)) #设置图片大小
plt.bar(x,y,color = "g") #绘制柱状图，表格给的数据是怎样就怎样，不会自动排序

plt.title("每个地区酒店数量", fontsize = 20) #设置标题
plt.xlabel("地区",fontsize = 18) 
plt.ylabel("酒店数量",fontsize = 18) #对横纵轴进行说明
plt.tick_params(labelsize = 10) #设置坐标轴上字体大小
plt.xticks(rotation = 45) #标签转90度

for a,b in zip(x,y): #数字直接显示在柱子上（添加文本）
    #a:要显示的值的x坐标，b:显示的值的y坐标，加上100是为了展示位置高一点点不重合，
    #第二个b:显示的文本的内容,ha,va:格式设定,center居中,top&bottom针对数字而言在上或者在下,fontsize:字体指定
    plt.text(a,b+3,b,ha = "center",va = "bottom",fontsize = 10) 

#plt.grid() #画网格线，有失美观因而注释点

plt.show()
```

（5）画出各个价格等级占比的饼图。

```python
y = data2.values
y = y/sum(y)

plt.figure(figsize = (7,7))
plt.title("价格等级占比",fontsize = 15)
patches,l_text,p_text = plt.pie(y, labels = data2.index, autopct = "%.1f %%", colors = "byg", startangle = 90,labeldistance = 1.1)

for i in p_text: #通过返回值设置饼图内部字体
    i.set_size(15)
    i.set_color('w')

for i in l_text: #通过返回值设置饼图外部字体
    i.set_size(15)
    i.set_color('r')
    
plt.legend() #图例
plt.show()
```
